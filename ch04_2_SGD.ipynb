## SGDClassifier
ÌôïÎ•†Ï†Å Í≤ΩÏÇ¨ ÌïòÍ∞ïÎ≤ïÏùÑ ÏÇ¨Ïö©Ìïú Î∂ÑÎ•ò Î™®Îç∏

```python
import pandas as pd 
fish = pd.read_csv('https://bit.ly/fish_csv_data')
```

```python
# speciesÏó¥(Ïª¨Îüº)ÏùÑ Ï†úÏô∏Ìïú ÎÇòÎ®∏ÏßÄ 5Í∞úÎäî ÏûÖÎ†• Îç∞Ïù¥ÌÑ∞Î°ú ÏÇ¨Ïö©
fish_input = fish[['Weight', 'Length', 'Diagonal', 'Height', 'Width']].to_numpy()
fish_target = fish['Species'].to_numpy()
```

```python
# ÌõàÎ†® ÏÑ∏Ìä∏ÏôÄ ÌÖåÏä§Ìä∏ ÏÑ∏Ìä∏Î°ú ÎÇòÎàÑÍ∏∞
from sklearn.model_selection import train_test_split
train_input, test_input, train_target, test_target = train_test_split(fish_input, fish_target, random_state=42)
```

```python
# ÌõàÎ†® ÏÑ∏Ìä∏ÏôÄ ÌÖåÏä§Ìä∏ ÏÑ∏Ìä∏Ïùò ÌäπÏÑ±ÏùÑ ÌëúÏ§ÄÌôî Ï†ÑÏ≤òÎ¶¨
# ÌõàÎ†® ÏÑ∏Ìä∏ÏóêÏÑú ÌïôÏäµÌïú ÌÜµÍ≥Ñ Í∞íÏúºÎ°ú ÌÖåÏä§Ìä∏ ÏÑ∏Ìä∏ÎèÑ Î≥ÄÌôòÌï¥Ïïº ÌïúÎã§
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
ss.fit(train_input)
train_scaled = ss.transform(train_input)
test_scaled = ss.transform(test_input)
```

```python
# ÌôïÎ•†Ï†Å Í≤ΩÏÇ¨ ÌïòÍ∞ïÎ≤ïÏùÑ Ï†úÍ≥µÌïòÎäî Î∂ÑÎ•òÏö© ÌÅ¥ÎûòÏä§ SGDClassifier
from sklearn.linear_model import SGDClassifier
```

```python
# 2Í∞úÏùò Îß§Í∞úÎ≥ÄÏàò ÏßÄÏ†ïÌïòÍ∏∞
# loss : ÏÜêÏã§ Ìï®ÏàòÏùò Ï¢ÖÎ•òÎ•º ÏßÄÏ†ï ; loss = 'log_loss' Î°úÏßÄÏä§Ìã± ÏÜêÏã§ Ìï®Ïàò ÏßÄÏ†ï
# max_iter : ÏàòÌñâÌï† ÏóêÌè¨ÌÅ¨ ÌöüÏàò ÏßÄÏ†ï 
sc = SGDClassifier(loss='log_loss', max_iter=10, random_state=42)
sc.fit(train_scaled, train_target)
print(sc.score(train_scaled, train_target))
print(sc.score(test_scaled, test_target))
```

### OvR(One versus Rest)
Îã§Ï§ë Î∂ÑÎ•òÏùº Í≤ΩÏö∞ loss='log_loss'Î°ú ÌïòÎ©¥ ÏßÄÏ†ïÎêú ÌÅ¥ÎûòÏä§ÎßàÎã§ Ïù¥ÏßÑ Î∂ÑÎ•ò Î™®Îç∏ÏùÑ ÎßåÎì†Îã§

ÎèÑÎØ∏Î•º ÏñëÏÑ±ÌÅ¥ÎûòÏä§Î°ú ÎëêÍ≥†, ÎÇòÎ®∏ÏßÄÎ•º Î™®Îëê ÏùåÏÑ± ÌÅ¥ÎûòÏä§Î°ú ÎëêÍ≥† Î™®Îç∏ÏùÑ ÎßåÎìúÎäî Ïãù

### partial_fit()
Î™®Îç∏ÏùÑ Ïù¥Ïñ¥ÏÑú ÌõàÎ†®Ìï† Îïå ÏÇ¨Ïö©ÌïòÎäî Î©îÏÑúÎìú

fit()ÏôÄ ÏÇ¨Ïö©Î≤ïÏù¥ Í∞ôÏßÄÎßå Ìò∏Ï∂úÌï† ÎïåÎßàÎã§ 1 ÏóêÌè¨ÌÅ¨Ïî© Ïù¥Ïñ¥ÏÑú ÌõàÎ†®Ìï† Ïàò ÏûàÎã§

```python
#  Ïù¥Ïñ¥ÏÑú ÌõàÎ†®ÌïòÍ∏∞
sc.partial_fit(train_scaled, train_target)
print(sc.score(train_scaled, train_target))
print(sc.score(test_scaled, test_target))
```

‚ùìtrain_scaled, train_targetÏùÑ ÌïúÍ∫ºÎ≤àÏóê ÏÇ¨Ïö©ÌñàÏúºÎãà ÌôïÎ•†Ï†Å Í≤ΩÏÇ¨ ÌïòÍ∞ïÎ≤ïÏù¥ ÏïÑÎãå Î∞∞Ïπò Í≤ΩÏÇ¨ ÌïòÍ∞ïÎ≤ïÏù¥ ÏïÑÎãåÍ∞Ä?
‚ùóÔ∏èÏïÑÎãàÎã§

    SGDClassifier Í∞ùÏ≤¥Ïóê ÌïúÎ≤àÏóê ÌõàÎ†® ÏÑ∏Ìä∏ Ï†ÑÏ≤¥Î•º Ï†ÑÎã¨ÌñàÏßÄÎßå, Ïù¥ ÏïåÍ≥†Î¶¨Ï¶òÏùÄ Ï†ÑÎã¨Ìïú ÌõàÎ†® ÏÑ∏Ìä∏ÏóêÏÑú 1Í∞úÏî© ÏÉòÌîåÏùÑ Í∫ºÎÇ¥ Í≤ΩÏÇ¨ ÌïòÍ∞ïÎ≤ï Îã®Í≥ÑÎ•º ÏàòÌñâÌïúÎã§!! 
    
    SGDClassifierÎäî Í≤ΩÏÇ¨ ÌïòÍ∞ïÎ≤ïÏù¥ÎÇò Î∞∞Ïπò ÌïòÍ∞ïÎ≤ïÏùÑ Ï†úÍ≥µÌïòÏßÄ ÏïäÎäîÎã§

## ÏóêÌè¨ÌÅ¨ÏôÄ Í≥ºÎåÄ /Í≥ºÏÜåÏ†ÅÌï©
ÌôïÎ•† Í≤ΩÏÇ¨ ÌïòÍ∞ïÎ≤ïÏùÑ ÏÇ¨Ïö©Ìïú Î™®Îç∏ÏùÄ ÌöüÏàòÏóê Îî∞Îùº Í≥ºÎåÄÏ†ÅÌï©Ïù¥ÎÇò Í≥ºÏÜåÏ†ÅÌï©Ïù¥ Îê† Ïàò ÏûàÎã§. 

üìçÏóêÌè¨ÌÅ¨ ÌöüÏàòÍ∞Ä Ï†ÅÏúºÎ©¥ Î™®Îç∏Ïù¥ ÌõàÎ†®ÏÑ∏Ìä∏Î•º Îçú ÌïôÏäµÌïúÎã§

    ÏÇ∞ÏùÑ Îã§ ÎÇ¥Î†§Ïò§ÏßÄ Î™ªÌïòÍ≥† ÌõàÎ†®ÏùÑ ÎßàÎ¨¥Î¶¨ÌïòÎäî ÏÖà = Í≥ºÏÜåÏ†ÅÌï©
    
üìçÏóêÌè¨ÌÅ¨ ÌöüÏàòÍ∞Ä ÎßéÏúºÎ©¥ ÌõàÎ†® ÏÑ∏Ìä∏Î•º ÏôÑÏ†ÑÌûà ÌïôÏäµÌïúÎã§

    ÌõàÎ†® ÏÑ∏Ìä∏Ïóê ÏïÑÏ£º Ïûò ÎßûÎäî Î™®Îç∏Ïù¥ ÎßåÎì§Ïñ¥ÏßÑÎã§ = Í≥ºÎåÄÏ†ÅÌï©



ÌõàÎ†® ÏÑ∏Ìä∏ Ï†êÏàòÎäî ÏóêÌè¨ÌÅ¨Í∞Ä ÏßÑÌñâÎê†ÏàòÎ°ù Íæ∏Ï§ÄÌûà Ï¶ùÍ∞ÄÌïòÎã§Í∞Ä Ïñ¥Îäê ÏàúÍ∞Ñ Í∞êÏÜåÌïòÍ∏∞ ÏãúÏûëÌïòÎäîÎç∞, Ïù¥ ÏßÄÏ†êÏù¥ Í≥ºÎåÄÏ†ÅÌï©Ïù¥ ÏãúÏûëÌïòÎäî Í≥≥

Í≥ºÎåÄÏ†ÅÌï©Ïù¥ ÏãúÏûëÌïòÍ∏∞ Ï†ÑÏóê ÌõàÎ†®ÏùÑ Î©àÏ∂îÎäî Í≤ÉÏùÑ Ï°∞Í∏∞Ï¢ÖÎ£å(early stopping)Ïù¥ÎùºÍ≥† ÌïúÎã§.

partial_fit() Î©îÏÑúÎìúÎßå ÏÇ¨Ïö©Ìï† Í≤ÉÏù∏Îç∞, Ìï¥Îãπ Î©îÏÑúÎìúÎßå ÏÇ¨Ïö©ÌïòÍ∏∞ ÏúÑÌï¥ÏÑúÎäî ÌõàÎ†® ÏÑ∏Ìä∏Ïóê ÏûàÎäî Ï†ÑÏ≤¥ ÌÅ¥ÎûòÏä§Ïùò Î†àÏù¥Î∏îÏùÑ Ï†ÑÎã¨Ìï¥ Ï£ºÏñ¥Ïïº ÌïúÎã§. 

‚û°Ô∏è np.unique() Ìï®ÏàòÎ°ú train_targetÏóê ÏûàÎäî 7Í∞ú ÏÉùÏÑ† Î™©Î°ùÏùÑ ÎßåÎì§Í≥†, ÏóêÌè¨ÌÅ¨ÎßàÎã§ ÌõàÎ†® ÏÑ∏Ìä∏ÏôÄ ÌÖåÏä§Ìä∏ ÏÑ∏Ìä∏Ïóê ÎåÄÌïú Ï†êÏàòÎ•º Í∏∞Î°ùÌïòÍ∏∞ ÏúÑÌï¥ 2Í∞úÏùò Î¶¨Ïä§Ìä∏Î•º ÎßåÎì§Í∏∞

```python
import numpy as np
sc  = SGDClassifier(loss='log_loss', random_state=42)
train_score = []
test_score = []
classes = np.unique(train_target)
```

```python
# 300Î≤àÏùò ÏóêÌè¨ÌÅ¨ ÎèôÏïà ÌõàÎ†®ÏùÑ Î∞òÎ≥µÌï¥ÏÑú ÏßÑÌñâÌïòÍ∏∞
# Îß§ Î∞òÎ≥µÎßàÎã§ ÌõàÎ†® ÏÑ∏Ìä∏ÏôÄ ÌÖåÏä§Ìä∏ ÏÑ∏Ìä∏Ïùò Ï†êÏàòÎ•º Í≥ÑÏÇ∞Ìï¥ Î¶¨Ïä§Ìä∏Ïóê Ï∂îÍ∞Ä
for _ in range(0, 300):
    sc.partial_fit(train_scaled, train_target, classes=classes)
    train_score.append(sc.score(train_scaled, train_target))
    test_score.append(sc.score(test_scaled, test_target))
```

ÌååÏù¥Ïç¨Ïùò `_`Îäî ÌäπÎ≥ÑÌïú Î≥ÄÏàò : ÎÇòÏ§ëÏóê ÏÇ¨Ïö©ÌïòÏßÄ ÏïäÍ≥† Í∑∏ÎÉ• Î≤ÑÎ¶¨Îäî Í∞íÏùÑ ÎÑ£Ïñ¥ÎëêÎäî Ïö©ÎèÑÎ°ú ÏÇ¨Ïö©ÌïúÎã§

```python
# 300Î≤àÏùò ÏóêÌè¨ÌÅ¨ ÎèôÏïà Í∏∞Î°ùÌïú ÌõàÎ†® ÏÑ∏Ìä∏ÏôÄ ÌÖåÏä§Ìä∏ ÏÑ∏Ìä∏Ïùò Ï†êÏàòÎ•º Í∑∏ÎûòÌîÑÎ°ú Í∑∏Î¶¨Í∏∞
import matplotlib.pyplot as plt
plt.plot(train_score)
plt.plot(test_score)
plt.xlabel('epoch')
plt.ylabel('accuracy')
plt.show()
```

- 100Î≤àÏß∏ ÏóêÌè¨ÌÅ¨ Ïù¥ÌõÑ ÌõàÎ†® ÏÑ∏Ìä∏ÏôÄ ÌÖåÏä§Ìä∏ ÏÑ∏Ìä∏Ïùò Ï†êÏàòÍ∞Ä Î≤åÏñ¥ÏßÑÎã§
- ÏóêÌè¨ÌÅ¨ Ï¥àÍ∏∞ÏóêÎäî Í≥ºÏÜåÏ†ÅÌï©ÎêòÏñ¥ ÌõàÎ†® ÏÑ∏Ìä∏ÏôÄ ÌÖåÏä§Ìä∏ ÏÑ∏Ìä∏Ïùò Ï†êÏàòÍ∞Ä ÎÇÆÎã§
- Î∞± Î≤àÏß∏ ÏóêÌè¨ÌÅ¨Í∞Ä Ï†ÅÏ†àÌïú Î∞òÎ≥µ ÌöüÏàòÎ°ú Î≥¥Ïù∏Îã§

```python
# Î∞òÎ≥µ ÌöüÏàòÎ•º 100Ïóê ÎßûÏ∂îÍ≥† Î™®Îç∏ÏùÑ Îã§Ïãú ÌõàÎ†®ÌïòÍ∏∞
# tol = None : ÏûêÎèôÏúºÎ°ú Î©àÏ∂îÏßÄ ÏïäÍ≥† max_iter=100ÍπåÏßÄ Î¨¥Ï°∞Í±¥ Î∞òÎ≥µ
sc = SGDClassifier(loss='log_loss', max_iter=100, tol=None, random_state=42)
sc.fit(train_scaled, train_target)
```

```python
print(sc.score(train_scaled, train_target))
print(sc.score(test_scaled, test_target))
```

```python
# ÌûåÏßÄ ÏÜêÏã§ ÏÇ¨Ïö©Ìï¥ Í∞ôÏùÄ Î∞òÎ≥µ ÌöüÏàò ÎèôÏïà Î™®Îç∏ ÌõàÎ†®ÌïòÍ∏∞
sc = SGDClassifier(loss='hinge', max_iter=100, tol=None, random_state=42)
sc.fit(train_scaled, train_target)
print(sc.score(train_scaled, train_target))
print(sc.score(test_scaled, test_target))
```
