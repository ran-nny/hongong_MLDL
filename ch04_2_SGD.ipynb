## 확률적 경사 하강법 (Stochastic Gradient Descent)
훈련 세트에서 샘플 하나씩 꺼내 손실 함수의 경사를 따라 최적의 모델을 찾는 알고리즘. 샘플을 여러 개 사용하면 미니배치 경사 하강법, 한 번에 전체 샘플을 사용하면 배치 경사 하강법

📍확률적 = 무작위하게, 랜덤하게

📍경사 : 기울기 (이 언덕은 경사가 참 가파르군!)

📍하강법 : 내려가는 방법 

➡️ 확률적 경사 하강법은 경사를 따라 내려가는 방법

산에서 내려온다면 가장 빠른 길을 선택하는 것이 좋은데, 가장 빠른 길은 경사가 가장 가파른 길이다. 

💡 경사 하강법 : 가장 가파른 경사를 따라 원하는 지점에 도달하는 것이 목표인데, 한 걸음이 너무 크면 경사를 따라 내려가지 못하고 오히려 올라갈 수 있다. 다라서 가장 가파른 길을 찾아 내려오지만 조금씩 내려오는 것이 중요하다.

💡 확률적 : 훈련 세트를 사용해 모델을 훈련하기 때문에 경사 하강법도 훈련 세트를 사용해 가장 가파른 길을 찾는데, 전체 샘플을 사용하지 않고 딱 하나의 샘플을 훈련 세트에서 랜덤하게 골라 가장 가파른 길을 찾는다.

➡️ 훈련 세트에서 랜덤하게 하나의 샘플을 고르는 것이 확률적 경사 하강법

    1. 훈련 세트에서 랜덤하게 하나의 샘플을 선택하여 가파른 경사를 조금씩 내려간다. 
    2. 그다음 훈련 센트에서 랜덤하게 또 다른 샘플을 하나 선택하여 경사를 조금 내려간다. 
    
❓ 모든 샘플을 다 사용했는데도 산을 내려오지 못한다면?

❗️ 다시 처음부터 시작. 즉 훈련 세트에 모든 샘플을 다시 채워 넣고 다시 랜덤하게 하나의 샘플을 이어서 경사를 내려간다. 

### 에포크(epoch) 
확률적 경사 하강법에서 훈련 세트를 한 번 모두 사용하는 과정

전체 샘플을 모두 사용하는 한 번 반복을 의미

    일반적으로 수십, 수백 번 이상의 에포크를 수행한다. 
    
### 미니배치 경사 하강법(minibatch gradient descent)
여러 개의 샘플을 사용해 경사 하강법을 수행하는 방식 - 실전에서 많이 사용

### 배치 경사 하강법(batch gradient descent)
한 번의 경사로를 따라 이동하기 위해 전체 샘플을 사용

📍전체 데이터를 사용하기 때문에 가장 안정적인 방법이지만, 전테 데이터를 사용하면 그만큼 컴퓨터 자원을 많이 사용하기 된다. 


📌 확률적 경사 하강법은 훈련 세트를 사용해 산 아래에 있는 최적의 장소로 조금씩 이동하는 알고리즘으로, 이로 인해 훈련 데이터가 모두 준비되어 있지 않고 매일매일 업데이트되어도 학습을 계속 이어나갈 수 있다. -> 다시 산꼭대기에서부터 시작할 필요가 없다.

📌 신경망알고리즘은 일반적으로 많은 데이터를 사용하기 때문에 한 번에 모든 데이터를 사용하기 어렵고, 모델이 매우 복잡해서 수학적인 방법으로 해답을 얻기 어렵기 때문에 확률적 경사하강법이나 미니배치 경사 하강법을 이용한다. 

## 손실 함수(loss function)
확률적 경사 하강법이 최적화할 대상. 

어떤 문제에서 머신러닝 알고리즘이 얼마나 엉터리인지 측정

- 손실 함수의 값이 작을수록 좋다
- 어떤 값이 최솟값인지는 모른다

    - 가능한 많이 찾아보고 만족할만한 수준이라면 산을 다 내려왔다고 인정해야 한다. 
    
➡️ 이 값을 찾아 조금씩 이동하려면 확률적 경사 하강법이 좋다! 

손실 함수는 문제마다 정해져 있다. 

## 비용 함수(cost function)
손실 함수의 다른말

- 손실 함수 : 샘플 하나에 대한 손실 
- 비용 함수 : 훈련  세트에 있는 모든 샘플에 대한 손실 함수의 합

보통은 엄격히 구분하지 않고 섞어서 사용

**분류에서의 손실 = 정답을 맞히지 못하는 것**


4개의 예측 중 2개만 맞았으므로 정확도는 0.5

❓정확도를 손실 함수로 사용할 수 있을까? 

    정확도에 음수를 취하면 -1.0이 가장 낮고 -0.0이 가장 높다. 
    
❗️정확도에는 치명적인 단점이 있다. 
- 4개의 샘플만 있다면 가능한 정확도는 0. 0.25. 0.5, 0.75, 1 이렇게 다섯 가지뿐이다. 
- 경사 하강법을 사용하면 아주 조금씩 내려오는데, 정확도가 이렇게 듬성듬성하면 경사 하강법을 이용해 조금씩 움직일 수 없다. 

    - 산의 경사면은 확실히 연속적이어야한다!! 
- 즉, 손실 함수는 미분 가능해야 한다. 

❓어떻게 연속적인 손실 함수를 만들 수 있을까?
- 로지스틱 회귀의 예측은 0 또는 1 이지만 확률은 0~1 사이의 어떤 값도 될 수 있으므로 연속적이다. 
- 샘플 4개의 예측 확률을 각각 0.9, 0.3, 0.2, 0.8 이라고 가정

### 로지스틱 손실 함수(logistic loss function)
### 이진 크로스엔트로피 손실 함수(binary cross-entropy loss function)
1. 첫 번재 샘플의 예측은 0.9이므로 양성 클래스의 타깃인 1과 곱한 다음 음수로 바꿀 수 있다. 

    - 예측이 1에 가까울수록 좋은 모델이다. 
    - 예측이 1에 가까울수록 예측과 타깃의 곱의 음수는 점점 작아진다



2. 두 번째 샘플의 예측은 0.3이다. 타깃이 양성 클래스(1)인데 거리가 멀다. 예측과 타깃을 곱해 음수로 바꿔 보면 -0.3
3. 세 번째 샘플의 타깃은 음성 클래스라 0이기 때문에 예측 확률인 0.2와 그대로 곱하면 안된다. 무조건 0이 될 것이기 때문 

    - 타깃을 마치 양성 클래스처럼 바꾸어 1로 만든 후 예측값도 양성 클래스에 대한 예측으로 바꾼다. 
    - 1-0.2 = 0.8로 사용하고, 타깃(1)과 곱해준 후 음수로 바꾸어준다.
    - 세 번째 샘플은 음성 클래스인 타깃을 맞추었기 때문에 손실이 낮아야 한다. -0.8은 꽤 낮은 손실이다
4. 네 번째 샘플도 타깃은 음성 클래스이지만, 정답을 맞히지는 못했다. 

    - 역시 타깃을 1로 바꾸고 예측 확률을 1에서  뺀 다음 곱해서 음수로 바꾸면 -0.2 
    - 네 번째 샘플의 손실이 높은 것을 확인할 수 있다.(값이 크다)


💡예측 확률에 로그 함수를 적용하면 좋다

    예측 확률의 범위는 0~1 사이인데, 로그 함수는 이 사이에서 음수가 되기 때문에 최종 손실 값은 양수가 된다. 
    
    손실이 양수가 되면 좀 더 이해하기 쉽다
    
    로그 함수는 0에 가까울수록 아주 큰 음수가 되기 때문에 손실을 아주 크게 만들어 모델에 큰 영향을 미칠 수 있다


- 양성 클래스(타깃 = 1)일 때 손실은 -log(예측 확률)로 계산한다
    - 확률이 1에서 멀어질수록 손실은 아주 큰 양수가 된다
- 음성 클래스(타깃 = 0)일 때 손실은 -log(1-예측 확률)로 계산한다
    - 예측 확률이 0에서 멀어질수록 손실은 아주 큰 양수가 된다
    
📌 이 손실 함수를 사용하면 로지스틱 회귀 모델이 만들어진다!! 

### 크로스엔트로피 손실 함수(cross-entropy loss function)

이진 분류는 로지스틱 손실 함수를 사용하고 다중 분류는 크로스엔트로피 손실 함수를 사용한다.

📌 회귀에서 사용하는 손실 함수

    평균 절댓값 오차, 평균 제곱 오차
    
    둘 다 값이 작을수록 좋은 모델

- 머신러닝 라이브러리가 손실 함수를 처리해주기 때문에 걱정할 필요가 없지만, 손실 함수가 무엇인지, 왜 정의를 해야 하는지 이해하는 것이 중요하다!

## SGDClassifier
확률적 경사 하강법을 사용한 분류 모델

```python
import pandas as pd 
fish = pd.read_csv('https://bit.ly/fish_csv_data')
```

```python
# species열(컬럼)을 제외한 나머지 5개는 입력 데이터로 사용
fish_input = fish[['Weight', 'Length', 'Diagonal', 'Height', 'Width']].to_numpy()
fish_target = fish['Species'].to_numpy()
```

```python
# 훈련 세트와 테스트 세트로 나누기
from sklearn.model_selection import train_test_split
train_input, test_input, train_target, test_target = train_test_split(fish_input, fish_target, random_state=42)
```

```python
# 훈련 세트와 테스트 세트의 특성을 표준화 전처리
# 훈련 세트에서 학습한 통계 값으로 테스트 세트도 변환해야 한다
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
ss.fit(train_input)
train_scaled = ss.transform(train_input)
test_scaled = ss.transform(test_input)
```

```python
# 확률적 경사 하강법을 제공하는 분류용 클래스 SGDClassifier
from sklearn.linear_model import SGDClassifier
```

```python
# 2개의 매개변수 지정하기
# loss : 손실 함수의 종류를 지정 ; loss = 'log_loss' 로지스틱 손실 함수 지정
# max_iter : 수행할 에포크 횟수 지정 
sc = SGDClassifier(loss='log_loss', max_iter=10, random_state=42)
sc.fit(train_scaled, train_target)
print(sc.score(train_scaled, train_target))
print(sc.score(test_scaled, test_target))
```

### OvR(One versus Rest)
다중 분류일 경우 loss='log_loss'로 하면 지정된 클래스마다 이진 분류 모델을 만든다

도미를 양성클래스로 두고, 나머지를 모두 음성 클래스로 두고 모델을 만드는 식

### partial_fit()
모델을 이어서 훈련할 때 사용하는 메서드

fit()와 사용법이 같지만 호출할 때마다 1 에포크씩 이어서 훈련할 수 있다

```python
#  이어서 훈련하기
sc.partial_fit(train_scaled, train_target)
print(sc.score(train_scaled, train_target))
print(sc.score(test_scaled, test_target))
```

❓train_scaled, train_target을 한꺼번에 사용했으니 확률적 경사 하강법이 아닌 배치 경사 하강법이 아닌가?
❗️아니다

    SGDClassifier 객체에 한번에 훈련 세트 전체를 전달했지만, 이 알고리즘은 전달한 훈련 세트에서 1개씩 샘플을 꺼내 경사 하강법 단계를 수행한다!! 
    
    SGDClassifier는 경사 하강법이나 배치 하강법을 제공하지 않는다

## 에포크와 과대 /과소적합
확률 경사 하강법을 사용한 모델은 횟수에 따라 과대적합이나 과소적합이 될 수 있다. 

📍에포크 횟수가 적으면 모델이 훈련세트를 덜 학습한다

    산을 다 내려오지 못하고 훈련을 마무리하는 셈 = 과소적합
    
📍에포크 횟수가 많으면 훈련 세트를 완전히 학습한다

    훈련 세트에 아주 잘 맞는 모델이 만들어진다 = 과대적합



훈련 세트 점수는 에포크가 진행될수록 꾸준히 증가하다가 어느 순간 감소하기 시작하는데, 이 지점이 과대적합이 시작하는 곳

과대적합이 시작하기 전에 훈련을 멈추는 것을 조기종료(early stopping)이라고 한다.

partial_fit() 메서드만 사용할 것인데, 해당 메서드만 사용하기 위해서는 훈련 세트에 있는 전체 클래스의 레이블을 전달해 주어야 한다. 

➡️ np.unique() 함수로 train_target에 있는 7개 생선 목록을 만들고, 에포크마다 훈련 세트와 테스트 세트에 대한 점수를 기록하기 위해 2개의 리스트를 만들기

```python
import numpy as np
sc  = SGDClassifier(loss='log_loss', random_state=42)
train_score = []
test_score = []
classes = np.unique(train_target)
```

```python
# 300번의 에포크 동안 훈련을 반복해서 진행하기
# 매 반복마다 훈련 세트와 테스트 세트의 점수를 계산해 리스트에 추가
for _ in range(0, 300):
    sc.partial_fit(train_scaled, train_target, classes=classes)
    train_score.append(sc.score(train_scaled, train_target))
    test_score.append(sc.score(test_scaled, test_target))
```

파이썬의 `_`는 특별한 변수 : 나중에 사용하지 않고 그냥 버리는 값을 넣어두는 용도로 사용한다

```python
# 300번의 에포크 동안 기록한 훈련 세트와 테스트 세트의 점수를 그래프로 그리기
import matplotlib.pyplot as plt
plt.plot(train_score)
plt.plot(test_score)
plt.xlabel('epoch')
plt.ylabel('accuracy')
plt.show()
```

- 100번째 에포크 이후 훈련 세트와 테스트 세트의 점수가 벌어진다
- 에포크 초기에는 과소적합되어 훈련 세트와 테스트 세트의 점수가 낮다
- 백 번째 에포크가 적절한 반복 횟수로 보인다

```python
# 반복 횟수를 100에 맞추고 모델을 다시 훈련하기
# tol = None : 자동으로 멈추지 않고 max_iter=100까지 무조건 반복
sc = SGDClassifier(loss='log_loss', max_iter=100, tol=None, random_state=42)
sc.fit(train_scaled, train_target)
```

```python
print(sc.score(train_scaled, train_target))
print(sc.score(test_scaled, test_target))
```

### 힌지 손실(hinge loss) 
loss 매개변수의 기본값은 hinge이다

힌지 손실은 서포트 벡터 머신(SVM)이라 불리는 또 다른 머신러닝 알고리즘을 위한 손실 함수

서포트 벡터 머신을 다루지는 않지만, 널리 사용되는 머신러닝 알고리즘 중 하나이다

DClassifier가 여러 종류의 손실 함수를 loss 매개변수에 지정하여 다양한 머신 러닝 알고리즘을 지원한다는 사실 기억하기

```python
# 힌지 손실 사용해 같은 반복 횟수 동안 모델 훈련하기
sc = SGDClassifier(loss='hinge', max_iter=100, tol=None, random_state=42)
sc.fit(train_scaled, train_target)
print(sc.score(train_scaled, train_target))
print(sc.score(test_scaled, test_target))
```

데이터의 규모가 커졌기 때문에 한 번에 컴퓨터 메모리에 읽을 수 없다. 

데이터를 조금씩 사용해 점진적으로 학습하는 방법이 필요해졌다 - 확률적 경사 하강법

## 핵심 패키지와 함수
### scikit-learn
- SGDClassifier : 확률적 경사 하강법을 사용한 분류 모델 만들기

    - loss : 확률적 경사 하강법으로 최적화할 손실 함수 지정
        - 기본값 : hinge손실함수 (SVM 위한)
    - penalty : 규제의 종류 지정
        - 기본값 : L2 규제를 위한 l2 
        - 규제 강도는 alpha 매개변수에 지정 (기본값 0.00001)
    - max_iter : 에포크 횟수를 지정
        - 기본값 : 1000
    - tol : 반복을 멈출 조건
        - n_iter_no_change 매개변수에서 지정한 에포크 동안 손실이 tol만큼 줄어들지 않으면 알고리즘이 중단된다
        - tol 기본값 0.001, n_iter_no_change 기본값 5
- SGDRegressor : 확률적 경사 하강법을 사용한 회귀 모델
    - loss매개변수에서 손실 함수 지정
        - 기본값 : squared_loss (제곱 오차)
    - SGDClassifier에서 설명한 매개변수는 모두 사용 가능
